## –†–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏ ‚Ññ1

```python
class WrongInputException(Exception):
    pass


def driver(input_func, process_func):
    data = input_func()
    res = process_func(*data)
    print(f"Result for {process_func.__name__} with {data} : {res}")


def input_list(size, left, rigth):
    a = [int(i) for i in input().split()]

    if len(a) != size:
        raise WrongInputException(f"Len of {a} must be equals {size}")

    for i in a:
        if i < left or i > rigth:
            raise WrongInputException(f"Element is not in given range")

    return a


def input_int(left, right):
    try:
        n = int(input())
        if n < left or n > right:
            raise WrongInputException(f"Size of number is not in given range")
        return n
    except ValueError:
        raise WrongInputException("Input a number")
```

```python
def input_task1_data():
    n = input_int(1, 100_000)
    m = input_int(1, 100_000)

    f = input_list(n, 1, n)
    b = input_list(m, 1, n)

    return f, b


def task1(f, b):
    s1 = set(f) 
    s2 = set(b) 

    ans = [] 

    if s2 <= s1: 
        d = {} 

        for i, fi in enumerate(f):
            if fi in d: 
                        
                d[fi].append(i + 1) 
            else: 
                d[fi] = [i + 1]

        for bi in b:
            if bi in d and len(d[bi]) > 1: 
                return 'Ambiguity'
            elif bi in d: 
                ans.append(d[bi][0])

        return 'Possible', ' '.join([str(i) for i in ans])

    return 'Impossible'


driver(input_task1_data, task1)
```

## –†–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏ ‚Ññ2

```python
def input_task2_data():
    n = input_int(1, 4000)
    t = input_list(n, 1, 106)
    return n, t

def task2(n, t):
    
    dp = [[0] * (n + 1) for _ in range(n + 1)]

    for i in range(n):
        p = -1
        for j in range(i):

            dp[i][j] = max(dp[i][j], dp[j][p] + 1)

            if t[j] == t[i]:
                p = j

    return max(max(dp[i]) for i in range(n)) + 1


driver(input_task2_data, task2)
```

## –†–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏ ‚Ññ3

```python
def input_task3_data():
    s = input()
    n = input_int(1, 105)


    letters = ("x", "y", "z")
    for letter in s:
        if letter not in letters:
            raise WrongInputException("All letters must be on of x, y, z")
        
    tests = []
    for _ in range(n):
        i = input().split()
        if len(i) != 2:
            raise WrongInputException()
            
        tests.append((int(i[0])-1, int(i[1])-1))
    return s, n, tests


def task3(s, n, tests):
    count_left = {}
    count_left[-1] = (0, 0, 0)
    x, y, z = 0, 0, 0

    for i, c in enumerate(s):
        if c == 'x':
            x += 1
        elif c == 'y':
            y += 1
        elif c == 'z':
            z += 1

        count_left[i] = (x, y, z)
    def count(a, b):
        x2, y2, z2 = count_left[b]
        x1, y1, z1 = count_left[a - 1]
        x = x2 - x1
        y = y2 - y1
        z = z2 - z1
        return x, y, z
     
    def is_valid(a, b):
        if b - a < 2: 
            return True
        x, y, z = count(a, b)
        diffs = [abs(i) for i in (x-y, x-z, y-z)] 
        return not any(d > 1 for d in diffs)
     
     
    result = []
    for ti, b in tests:
        result.append("YES" if is_valid(ti, b) else "NO")

    return ' '.join(result)


driver(input_task3_data, task3)
```

## –†–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏ ‚Ññ4

```python
def input_task4_data():
    n = input_int(1, 105)
    t = input_list(n, 0, n-1)
    return n, t


def task4(n, t):

    s = sorted([(n-ti, i) for i, ti in enumerate(t)])

    ans = []

    hat_type = 0 
    one_type_hats_processed_count = 0 

    for i in range(n):
        if one_type_hats_processed_count == 0:
            hat_type += 1
            ans.append((s[i][1], hat_type))
            one_type_hats_processed_count = s[i][0]-1
        else:
            if s[i][0] == s[i-1][0]:
                one_type_hats_processed_count -= 1
                ans.append((s[i][1], hat_type))
            else:
                return 'Impossible'

    if hat_type > n or one_type_hats_processed_count > 0:
        return 'Impossible'

    else:
        ans.sort()
        return 'Possible', [i[1] for i in sorted(ans)]

driver(input_task4_data, task4)
```

## –†–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏ ‚Ññ5

```python
def input_task5_data():
    n = input_int(1, 100_000)
    p = input_list(n, 1, n)
    return (p, )


def task5(t):
    p = [0] + t
    val = sorted([dfs(u, p) for u in p])
    return sum([x*x for x in val]) + 2 * val[-1] * val[-2]

def dfs(u, p):
  res = 0
  while p[u] != 0:
    res += 1
    p[u], u = 0, p[u]

  return res
  
driver(input_task5_data, task5)
```

## –†–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏ ‚Ññ6

```python
def input_task6_data():
    n = input_int(1, 1000)
    l = input_list(n, 0, n)
    r = input_list(n, 0, n)
    return n, l, r


def task6(n, l, r):
    candies = [] 
    i = 0 
    l1 = []
    r1 = []

    for i in range(n):
        if l[i] > i or r[i] > n - i - 1:
            print('NO')
            break

        candies.append(n - l[i] - r[i])

    for i in range(n):
        for i in range(i):
            if candies[i] < candies[i]:
                i += 1
        l1.append(i)
        i = 0

        for i in range(i + 1, n):
            if candies[i] < candies[i]:
                i += 1
        r1.append(i)
        i = 0

    if l1 == l and r1 == r:
        return "YES", candies

    return "NO"


driver(input_task6_data, task6)
```

## –†–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏ ‚Ññ7

```python
def input_task7_data():
    n = input_int(1, 105)
    k = input_int(1, 4)
    a = [
        input_list(k, 0, 1)
        for _ in range(n)
    ]
    return k, a


def task7(k, a):
    sq_k = 2 << (k - 1) 
    rec = [0] * sq_k 

    for i in a:
        ind = int(''.join(i), 2) 
        rec[ind] += 1

    ans = 'YES'
    if rec[0] > 0:
        return ans
    for k in range(sq_k):
        for k in range(sq_k):
            if k & k == 0 and rec[k] > 0 and rec[k] > 0:
                return 'YES'

    return 'NO'


driver(input_task7_data, task7)
```

## –†–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏ ‚Ññ8

```python
def input_task8_data():
    t = input()

    if len(t) > 2000:
        raise WrongInputException(f"String is too long")

    return (t, )

def task8(t):
    t_size = len(t)
    i = t_size // 2 
    n, a = [1] * t_size, [1] * t_size

    for x, y in [(i, i + 2) for i in range(i - 1)]:
        while x >= 0 and t[x] == t[y]:
            n[y] += 1
            a[x] += 1
            x -= 1
            y += 1

    for x, y in [(i, i + 2) for i in range(i - 1, t_size - 1)]:
        while y < t_size and t[x] == t[y]:
            n[y] += 1
            a[x] += 1
            x -= 1
            y += 1

    for x, y in [(i, i + 1) for i in range(i)]:
        while x >= 0 and t[x] == t[y]:
            n[y] += 1
            a[x] += 1
            x -= 1
            y += 1
    for x, y in [(i, i + 1) for i in range(i, t_size)]:
        while y < t_size and t[x] == t[y]:
            n[y] += 1
            a[x] += 1
            x -= 1
            y += 1

    for i in range(t_size - 1):
        n[i + 1] += n[i]

    return sum(n[i] * a[i + 1] for i in range(t_size - 1))

driver(input_task8_data, task8)
```

## –†–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏ ‚Ññ9

```python

def input_task9_data():
    n, m = input().split()
    try:
        n = int(n)
        m = int(m)
    except ValueError:
        raise WrongInputException(f"Wrong input format")

    if n < 1 or m < 1 or n > 20 or m > 20:
        raise WrongInputException(f"Wrong range of values")

    a = [
        input_list(m, 0, m)
        for _ in range(n)
    ]

    return n, m, a


def task9(n, m, a):
    for i in range(m): 
        for j in range(i, m): 
            cnt = 0 

            for row in range(n):
                a[row][i], a[row][j] = a[row][j], a[row][i]
                for k in range(m):
                    if a[row][k] != k + 1:
                        cnt += 1
            if cnt < 3:
                return "YES"

            for row in range(n):
                a[row][i], a[row][j] = a[row][j], a[row][i]
     
    return "NO"

driver(input_task9_data, task9)
```

```python
def input_task10_data():
    n = input_int(2, 105)
    a = input_list(n-1, 0, 3)
    b = input_list(n-1, 0, 3)

    return n, a, b


def task10(n, a, b):
    t = [5] * n 

    for i in range(4):
        t[0] = i 

        for j in range(n-1):
            for k in range(4):
                if t[j] | k == a[j] and t[j] & k == b[j]:
                    t[j+1] = k
        if 5 not in t:
            return "YES", t

    return "NO"

driver(input_task10_data, task10)
```
